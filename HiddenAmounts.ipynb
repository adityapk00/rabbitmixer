```python
import ecdsa
from ecdsa.ellipticcurve import Point
```

```python
abp = 1
bbp = 2
t = 5
```

```python
G = ecdsa.SECP256k1.generator
T = t * G
Abp = abp *  G
Bbp = bbp * G

nT =  Point(ecdsa.SECP256k1.curve, T.x(), -T.y() % ecdsa.SECP256k1.curve.p())

Abn = Abp + nT
Bbn = Bbp + T

abn = abp - t
bbn = bbp + t

abG = (abp - abn) * G
bbG = (bbn - bbp) * G

abpG = abp * G
abnG = abn * G

bbnG = bbn * G
bbpG = bbp * G

print(abnG)
print(Abn)


print(abpG)
print(Abp)


print(bbnG)
print(Bbn)


print(bbpG)
print(Bbp)
```
